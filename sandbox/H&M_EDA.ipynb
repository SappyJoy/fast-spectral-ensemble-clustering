```python
data = transactions.merge(articles, on='article_id', how='left')
data = data.drop('detail_desc', axis=1)
```

```python
# Проверка пропусков
print(data.isnull().sum())
```

```python
categorical_cols_to_drop = [
    'product_type_name', 'graphical_appearance_name', 'colour_group_name', 
    'perceived_colour_value_name', 'perceived_colour_master_name', 
    'department_name', 'index_name', 'index_group_name', 
    'section_name', 'garment_group_name', 'prod_name'
]

data = data.drop(categorical_cols_to_drop, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols_to_encode = [
    'product_group_name', 'index_code'
]

encoder = LabelEncoder()
# Создание датафрейма для хранения закодированных признаков
encoded_features = pd.DataFrame()

# Кодирование каждого категориального столбца по отдельности
for col in categorical_cols_to_encode:
    encoded_features[col] = encoder.fit_transform(data[col])

# Удаление оригинальных категориальных столбцов
data.drop(columns=categorical_cols_to_encode, inplace=True)

# Объединение с исходными данными
data = pd.concat([data.reset_index(drop=True), encoded_features.reset_index(drop=True)], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['price', 'sales_channel_id', 'product_code', 'product_type_no', 'graphical_appearance_no', 'colour_group_code', 'perceived_colour_value_id', 'perceived_colour_master_id', 'department_no', 'index_code', 'index_group_no', 'section_no', 'garment_group_no']

scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```
